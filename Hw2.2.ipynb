```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 1. Reading file
df = pd.read_csv('./2017_jun_final.csv')
```

```python
# 2. Displaying the first few rows of the DataFrame
df.head()
```

```python
## 3. Determine the size of the table using the shape method.
df.shape
```

```python
## 4. Determine the types of all columns using dataframe.dtypes.
df.dtypes
```

```python
## 5. Calculate the proportion of missing values in each column (use the isnull and sum methods).
calculate_null = df.isnull().sum()
print("Result:\n", calculate_null)
```

```python
## 6.1 Remove all columns with missing values except for the column "Programming Language".
tmp = df[df.columns.difference(['Язык.программирования'])].isna().any()
df.drop(tmp.index[tmp], axis=1, inplace=True)
```

```python
## 6.2 Again, calculate the proportion of missing values in each column and ensure that only the column "Programming Language" remains.
calculate_null = df.isnull().sum()
print("Result:\n", calculate_null)
```

```python
## 7. Remove all rows in the original table using the dropna method.
df.dropna(inplace=True)
```

```python
## 8. Determine the new size of the table using the shape method.
df.shape
```

```python
## 9. Create a new table 'python_data', which will contain only rows with specialists who specified Python as their programming language.
python_data = df[df["Язык.программирования"] == "Python"]
python_data
```

```python
## 10. Determine the size of the python_data table using the shape method.
python_data.shape
```

```python
## 11. Using the 'groupby' method, perform grouping by the column "Position".
grouped_python_data = python_data.groupby("Должность")
```

```python
## 12. Create a new DataFrame where, for the data grouped by the column "Position", perform data aggregation using the agg method and find the minimum and maximum values in the column "Salary.per.Month".
salary = grouped_python_data.agg({"Зарплата.в.месяц": ["min", "max"]})
salary
```

```python
## 13. Create a function named fill_avg_salary that will return the average monthly salary value. Use this function with the apply method to create a new column named "avg".
def fill_avg_salary(row):
    return row.mean()
salary[('Зарплата.в.месяц', 'avg')] = salary.apply(fill_avg_salary, axis=1)
salary
```

```python
## 14. Create descriptive statistics using the describe method for the new column.
descriptive_statistics = salary['Зарплата.в.месяц', 'avg'].describe()
descriptive_statistics
```

```python
## 15. Save the resulting table to a CSV file.
salary.to_csv('salary.csv')
grouped_df = df.groupby(['Язык.программирования', 'Пол'])['Зарплата.в.месяц'].mean().unstack()
```

```python
# Bar chart
plt.figure(figsize=(8, 6))  
grouped_df.plot(kind='bar', ax=plt.gca())
plt.xlabel('Programming Language')
plt.ylabel('Average salary per month')
plt.title('Distribution of average salaries by programming languages and gender')
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend(title='Gender')
plt.show()
```

```python
# bar chart
grouped_df = df.groupby(['Язык.программирования', 'Должность'])['Зарплата.в.месяц'].mean().unstack()
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar', ax=plt.gca())
plt.hist(grouped_df, bins=2, alpha=0.5)
plt.title('Distribution of average salaries by programming languages and position')
plt.xlabel('Programming Language')
plt.ylabel('Average salary per month')
plt.xticks(rotation=90)
plt.legend(title='Position')
plt.show()
```

```python
# distribution schedule
grouped_df = df.groupby(['Возраст', 'Уровень.английского']).size().unstack().fillna(0)

plt.figure(figsize=(10, 6))
grouped_df.plot(kind='line', ax=plt.gca(), stacked=True)
plt.xlabel('Age')
plt.ylabel('Number of candidates')
plt.title('Breakdown of candidates by English level and age')
plt.tight_layout()
plt.legend(title='English level')
plt.show()
```
